```python
import zipfile
import os

with zipfile.ZipFile("video-games-sale.zip", 'r') as zip_ref:
    zip_ref.extractall("video_games_data")

# Check the files
os.listdir("video_games_data")
```

```python
import pandas as pd

df = pd.read_csv("video_games_data/video games sales.csv")
df.head()
```

```python
df.columns
```

```python
# Sort dataset by Global_Sales descending
top10 = df.sort_values(by="Global_Sales", ascending=False).head(10)

# Select relevant columns
top10 = top10[["Name", "NA_Sales", "EU_Sales", "JP_Sales", "Global_Sales"]]

top10
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 7))

# Plot stacked bars
plt.bar(top10["Name"], top10["NA_Sales"], label="North America")
plt.bar(top10["Name"], top10["EU_Sales"], bottom=top10["NA_Sales"], label="Europe")
plt.bar(top10["Name"], top10["JP_Sales"], bottom=top10["NA_Sales"]+top10["EU_Sales"], label="Japan")

plt.xticks(rotation=45, ha="right")
plt.xlabel("Game Title")
plt.ylabel("Sales (millions)")
plt.title("Top 10 Video Games by Sales (NA, EU, JP)")
plt.legend()
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sort top10 by Global_Sales descending
top10_sorted = top10.sort_values(by="Global_Sales", ascending=False)

plt.figure(figsize=(12, 7))

# Plot stacked bars
bars_na = plt.bar(top10_sorted["Name"], top10_sorted["NA_Sales"], label="North America")
bars_eu = plt.bar(top10_sorted["Name"], top10_sorted["EU_Sales"], bottom=top10_sorted["NA_Sales"], label="Europe")
bars_jp = plt.bar(top10_sorted["Name"], top10_sorted["JP_Sales"], bottom=top10_sorted["NA_Sales"]+top10_sorted["EU_Sales"], label="Japan")

# Add total sales numbers above each bar with a dynamic offset
for i, total in enumerate(top10_sorted["Global_Sales"]):
    offset = total * 0.01  # 1% above the bar
    plt.text(i, total + offset, f"{total:.2f}", ha="center", va="bottom", fontsize=9, fontweight="bold")

plt.xticks(rotation=45, ha="right")
plt.xlabel("Game Title")
plt.ylabel("Sales (millions)")
plt.title("Top 10 Video Games by Sales (NA, EU, JP)")
plt.legend()
plt.tight_layout()

plt.show()
```
